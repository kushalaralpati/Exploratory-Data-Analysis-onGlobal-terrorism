# Data Wrangling Code

```python
# Write your code to make your dataset analysis ready.
gt_df.drop(columns={'eventid','extended','iday','country','targsubtype1','region','weaptype1','weapsubtype1','specificity',
                             'vicinity','crit1','crit2','crit3','doubtterr','multiple','suicide','attacktype1','targtype1','targsubtype1_txt'
                             ,'target1','natlty1','guncertain1','individual','weapsubtype1_txt','property','dbsource','INT_LOG','INT_IDEO',
                             'INT_MISC','INT_ANY'},inplace=True,axis=1)
```

```python
gt_df.columns
```

```python
gt_df['State']= gt_df['State'].fillna('Unknown')
gt_df['City']=gt_df['City'].fillna('Unknown')
gt_df['Target_type']= gt_df['Target_type'].fillna("Others")
gt_df['Nationality']=gt_df['Nationality'].fillna('Unknown')
gt_df['Kills']=gt_df['Kills'].fillna(0)
gt_df['Wounds']=gt_df['Wounds'].fillna(0)
gt_df['victims']=gt_df['Kills']+gt_df['Wounds']
```

#### Chart - 3 KDE Plot of Demographical Comparision Of Casuality Types Over The Year. - (Bi-Variate)

```python
# Chart - 3 visualization code
# Pair Plot visualization code

import matplotlib.pyplot as plt
import seaborn as sns

# Prepare DF
year_kills = gt_df.groupby('Year')['Kills'].sum()
year_kills = year_kills.drop('Unknown', errors='ignore')

year_wound = gt_df.groupby('Year')['Wounds'].sum()
year_wound = year_wound.drop('Unknown', errors='ignore')

# Set up the figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7))

# Plot the KDE plot for year_kills
sns.kdeplot(data=year_kills, x=year_kills.index, y=year_kills.values, fill=True, cmap='Reds', ax=ax1)
ax1.set(xlabel='Year', ylabel='Casualties(Killed)', title='Density Plot: Casualities(Killed) Over The Year\n')
ax1.set_xlim(1971, 2020)
ax1.set_ylim(0, 50000)

# Plot the KDE plot for year_wound
sns.kdeplot(data=year_wound, x=year_wound.index, y=year_wound.values, fill=True, cmap='Blues', ax=ax2)
ax2.set(xlabel='Year', ylabel='Casualties(Wounded)', title='Density Plot: Casualities(Wounded) Over The Year\n')
ax2.set_xlim(1971, 2020)
ax2.set_ylim(0, 50000)

# Adjust spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

##### 1.Why did you pick the specific chart?

A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analogous to a histogram. KDE represents the data using a continuous probability density curve in one or more dimensions.

#### Chart - 11 Analysis Of Terrorist Activities In The World : (Multi-Variate)

```python
# Chart - 11 visualization code
#Creating custom function to tabulate attack failures -
def failure_count(num):
    if num == 0:
        return 1
    else:
        return 0
```

```python
#Code -
var = gt_df
attack_group_w = var[['Year','Month','Success','Gang_name']]

attack_group_w.insert(loc = 4, column = 'Failure',value = attack_group_w['Success'].apply(failure_count))
overall = pd.pivot_table(data = attack_group_w,index = 'Gang_name',values = ['Success','Failure'],aggfunc = np.sum)
overall['Total_attacks'] = overall['Failure']+overall['Success']
overall['Success_rate'] = round((overall['Success']/overall['Total_attacks'])*100,2)

overall_top10 = overall.nlargest(n = 11, columns= 'Total_attacks')
overall_top10 = overall_top10.drop('Unknown',axis = 0)

overall_top5 = overall_top10.nlargest(5,columns= 'Total_attacks')
top5gang = overall_top5.reset_index()['Gang_name']
```
